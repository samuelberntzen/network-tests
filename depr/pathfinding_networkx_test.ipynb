```python
place = "Oslo"

import json
import os
import geopandas as gpd


os.environ["USE_PYGEOS"] = "0"
os.environ["LD_LIBRARY_PATH"] = "/usr/lib/x86_64-linux-gnu/libstdc++.so.6"

# Specify the file paths
nodes_filepath = f"data/{place}_nodes.geojson"
edges_filepath = f"data/{place}_edges.geojson"

# Load GeoJSON files as GeoDataFrames
nodes_gdf = gpd.read_file(nodes_filepath)
edges_gdf = gpd.read_file(edges_filepath)
```

```python
nodes_gdf["x"] = nodes_gdf["geometry"].x
nodes_gdf["y"] = nodes_gdf["geometry"].y
```
